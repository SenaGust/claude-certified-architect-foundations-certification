```python
import os
from dotenv import load_dotenv
from anthropic import Anthropic

load_dotenv()

auth_token = os.getenv("ANTHROPIC_API_KEY")

client = Anthropic(
    auth_token=auth_token,
    base_url="https://flow.ciandt.com/flow-llm-proxy/"
)

model = "bedrock/anthropic.claude-4-6-sonnet"

def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages,
        system="You are expert python engineer who writes very concise code."
    )
    return message.content[0].text

# Start with an empty message list
messages = []
 
user_input = "You: Write a Python function that checks a string for duplicated characters."
add_user_message(messages, user_input)
claude_answer = chat(messages)
print("Claude: " + claude_answer)
add_assistant_message(messages, claude_answer)
```
